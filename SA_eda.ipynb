```python
# Top n by X - get % of total Y
```

```python
# Function for TopN ( Takes inout as df, col on  which we want top n, col to aggregate, and the top n number)
# Returns the topn data frame
def func_calc_topN(df,groupByCol,aggCol,topN):
    dt = df.groupby(groupByCol,as_index=False)[aggCol].sum()
    dt = dt.sort_values([aggCol],ascending=False)
    dt.reset_index(drop=True, inplace=True)
    dt['Percent'] = (dt[aggCol]/dt[aggCol].sum()) * 100
    dt['Percent_cum_sum'] = dt['Percent'].cumsum()
    return(dt.head(topN))
```

```python
dt = func_calc_topN(df,'Short Text','Gross value',24)
dt
```
